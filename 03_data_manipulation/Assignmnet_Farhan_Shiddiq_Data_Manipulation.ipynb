**4. Pada waktu apa transaksi paling ramai (pagi, siang, sore, malam)?**

```python
# Mengkategirikan Time -> Pagi, Siang, Sore, Malam
df_supermarket_sales['Time'] = pd.to_datetime(df_supermarket_sales['Time'], format='%H:%M')
df_supermarket_sales['Hour'] = df_supermarket_sales['Time'].dt.hour

def get_time_of_day(hour):
    if 05.00 <= hour < 12.00:
        return 'Pagi'
    elif 12.00 <= hour < 17.00:
        return 'Siang'
    elif 17.00 <= hour < 20.00:
        return 'Sore'
    else:
        return 'Malam'

df_supermarket_sales['Waktu'] = df_supermarket_sales['Hour'].apply(get_time_of_day)
display(df_supermarket_sales[['Time', 'Hour', 'Waktu']].head())
```

```python
# Menampilkan banyaknya payment berdasarkan waktu
transaksi_per_waktu = df_supermarket_sales.groupby('Waktu')['Invoice ID'].count().sort_values(ascending=False)
display(transaksi_per_waktu)
```

**INSIGHT**

Berdasarkan hasil pengelompokan waktu transaksi menjadi pagi, siang, sore, dan malam, terlihat bahwa transaksi paling banyak terjadi pada waktu siang. Hal ini menunjukkan bahwa sebagian besar pelanggan melakukan pembelian pada jam tengah hari hingga sore hari, kemungkinan karena bertepatan dengan waktu istirahat kerja atau aktivitas belanja setelah makan siang.

Transaksi pada waktu pagi juga cukup banyak, yang mengindikasikan bahwa sebagian pelanggan melakukan belanja lebih awal, mungkin sebelum memulai aktivitas harian mereka. Sementara itu, jumlah transaksi pada waktu sore dan malam cenderung lebih sedikit dibandingkan waktu siang, meskipun masih terdapat pelanggan yang berbelanja pada waktu tersebut.

Pola ini menunjukkan bahwa aktivitas belanja pelanggan di supermarket cenderung meningkat pada siang hari, sehingga waktu tersebut dapat dianggap sebagai jam sibuk (peak hours) bagi supermarket.

# 2. Standar Scaller for Exploratory Data Analysis ?
- lakukan pada kolom Total, Unit Price, dan tambahkan 2 kolom lainnya
- dan tulis insight dari yang anda dapat dari grafik tersebut

```python
# Melakukan standard scaling
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

cols = ['Total','Unit price','Quantity','Rating']

scaler = StandardScaler()

scaled = scaler.fit_transform(df_supermarket_sales[cols])

df_scaled = pd.DataFrame(scaled, columns=cols)
```

```python
# Visualisasikan menggunakan grafik

import seaborn as sns
import matplotlib.pyplot as plt

cols = ['Total','Unit price','Quantity','Rating']

plt.figure(figsize=(12,8))

for i, col in enumerate(cols, 1):

    plt.subplot(2,2,i)

    sns.histplot(df_scaled[col], bins=30, kde=True) #type: ignore

    plt.title(f'Distribusi {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

+ Setelah dilakukan standardisasi, semua variabel (Total, Unit price, Quantity, Rating) terpusat di sekitar nilai 0, yang menunjukkan proses scaling berhasil menormalkan data.

+ Skala antar variabel menjadi seragam, sehingga perbandingan distribusi setiap variabel lebih mudah dilakukan.

+ Sebagian besar data berada di sekitar nilai rata-rata, menandakan mayoritas transaksi tidak terlalu jauh dari pola transaksi umum.

+ Beberapa nilai berada cukup jauh dari pusat distribusi, yang menunjukkan kemungkinan adanya transaksi ekstrem (outlier).


**Kesimpulan: StandardScaler membantu menormalkan skala data sehingga distribusi variabel lebih mudah dianalisis dan dibandingkan.**

# 3. MinMax Scaller for Exploratory Data Analysis ?
- lakukan pada kolom Total, Unit Price, dan tambahkan 2 kolom lainnya
- dan tulis insight dari yang anda dapat dari grafik tersebut

```python
from sklearn.preprocessing import MinMaxScaler

# Kolom-kolom yang ingin dinormalisasi
columns_to_minmax = ['Total', 'Unit price', 'Quantity', 'Rating']

# Inisialisasi MinMaxScaler
minmax = MinMaxScaler()
```

```python
# Menerapkan normalisasi Min-Max
df_supermarket_sales[['Total MinMax', 'Unit Price MinMax', 'Quantity MinMax', 'Rating MinMax']] = minmax.fit_transform(df_supermarket_sales[columns_to_minmax])
```

```python
# Checking result of standar scaller process
df_supermarket_sales[['Total MinMax', 'Unit Price MinMax', 'Quantity MinMax', 'Rating MinMax']].head()
```

```python
# After standar scaller process the result has -> min (minimum) = 0 and max (maximum) = 1
df_supermarket_sales[['Total MinMax', 'Unit Price MinMax', 'Quantity MinMax', 'Rating MinMax']].describe()
```

```python
# Visualisasikan menggunakan grafik

import seaborn as sns
import matplotlib.pyplot as plt

cols = ['Total MinMax', 'Unit Price MinMax', 'Quantity MinMax', 'Rating MinMax']

plt.figure(figsize=(12,8))

for i, col in enumerate(cols, 1):

    plt.subplot(2,2,i)

    sns.histplot(df_supermarket_sales[col], bins=30, kde=True) #type: ignore

    plt.title(f'Distribusi {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

1. Total MinMax
    
    Distribusi cenderung miring ke kanan (right skewed). Artinya sebagian besar transaksi memiliki nilai total yang relatif kecil, sementara hanya sedikit transaksi dengan total yang sangat besar.

2. Unit Price MinMax
    
    Distribusi harga produk terlihat cukup merata di seluruh rentang 0–1, menunjukkan variasi harga produk di supermarket cukup beragam.

3. Quantity MinMax
    
    Distribusi terlihat terpisah-pisah (diskrit) karena quantity merupakan jumlah barang yang dibeli, sehingga nilainya terbatas pada angka tertentu.

4. Rating MinMax
    
    Distribusi rating relatif menyebar merata, yang menunjukkan bahwa penilaian pelanggan cukup bervariasi dan tidak hanya terpusat pada satu nilai saja.

**Kesimpulan**

Setelah menggunakan MinMaxScaler, semua variabel berada pada rentang 0-1, sehingga perbandingan distribusi antar variabel menjadi lebih mudah tanpa mengubah bentuk distribusi aslinya.
